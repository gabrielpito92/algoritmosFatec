1. Criar um algoritmo que armazene números em dois vetores inteiros de cinco elementos cada. Gere e imprima o vetor soma.

image.png

```python
# O resultado é o mesmo do exercício proposto
num_vectors = 2
num_rows = 1
num_cols = 5
size_v = num_rows * num_cols

vectors = create_vectors(num_vectors, size_v)

for i in range(num_vectors):
    print(f'\nVocê irá entrar com os valores do vetor {i + 1}\n')
    for j in range(num_cols):
        vectors[i][j] = int(input(f'Digite o {j + 1}° valor do vetor: '))

print('\nSoma das colunas:\n')

column_sum = np.sum(vectors, axis=0)
for c, v in enumerate(column_sum):
    print(f'soma[{c + 1}]: {int(v)}')
```

2. Ler um vetor vet de 10 elementos e obter um vetor w cujos componentes são os fatoriais dos respectivos componentes de v.

```python
num_vectors = 1
num_rows = 1
num_cols = 10
size_v = num_rows * num_cols

vectors_2 = create_vectors(num_vectors, size_v) #cria vetor pela função anterior

for i in range(num_vectors):
    print(f'\nVocê irá entrar com os valores do vetor {i + 1}\n')
    for j in range(num_cols):
        num_fact = int(input(f'Digite o {j + 1}° valor do vetor: '))
        result = str(np.math.factorial(num_fact))
        vectors_2[i][j] = result

print('\n--- Resultados Fatoriais ---')
for n in range(num_vectors):
    print(f'\nFatoriais do vetor {n + 1}\n')

    for ind, val in enumerate(vectors_2[n]):
        print(f'{ind + 1}° Fatorial: {int(val)}')
```

3. Entrar com números inteiros em um vetor A [10]. Gerar e imprimir o vetor B onde cada elemento é o quadrado do elemento, na respectiva posição, do vetor A.

```python
num_vectors = 2
num_rows = 1
num_cols = 10
size_v = num_rows * num_cols

# Criando vetor A e B (2 vetores)
vectors_3 = create_vectors(num_vectors, size_v) #cria vetor pela função anterior

for a in range(size_v):
    num = int(input(f'Digite o {a + 1}° valor do vetor: '))
    result = pow(num, 2)
    vectors_3[0][a] = num
    vectors_3[1][a] = result

print('\n')
printVecs(vectors_3)
```

1. Criar um algoritmo que leia os elementos de uma matriz inteira 5x5 e escreva os elementos da diagonal principal, acima da diagonal principal e abaixo da diagonal principal.

```python
num_vectors = 1
size_matriz_quad = 5
size_v = [5,5]

vectors_6 = create_vectors(num_vectors, size_v) # Criando matriz
vectors_6 = vectors_6[0]

print("Digite os elementos da matriz 5x5:") # Preencha a matriz
for i in range(size_matriz_quad):
  print('\n')

  for j in range(size_matriz_quad):
    vectors_6[i][j] = int(input(f"Digite o elemento [{i}][{j}]: "))

print("\nElementos da diagonal principal:") # diagonal principal
for i in range(size_matriz_quad):
  print(int(vectors_6[i][i]), end=" ")

print("\n\nElementos acima da diagonal principal:") # acima da diagonal principal
for i in range(size_matriz_quad):
  for j in range(i + 1, size_matriz_quad):
    print(int(vectors_6[i][j]), end=" ")


print("\n\nElementos abaixo da diagonal principal:") # abaixo da diagonal principal
for i in range(size_matriz_quad):
  for j in range(i):
    print(int(vectors_6[i][j]), end=" ")
```

2. Criar um algoritmo que leia os elementos de uma matriz inteira 5x5 e escreva os elementos da diagonal secundária, acima da diagonal secundária e abaixo da diagonal secundária.

```python
num_vectors = 1
size_matriz_quad = 5
size_v = [5,5]

vectors_7 = create_vectors(num_vectors, size_v) # Criando matriz
vectors_7 = vectors_7[0]

print("Digite os elementos da matriz 5x5:") # Preencher a matriz
for i in range(size_matriz_quad):
  print('\n')

  for j in range(size_matriz_quad):
    vectors_7[i][j] = int(input(f"Digite o elemento [{i}][{j}]: "))

print("\nElementos da diagonal secundária:") # El. diagonal secundária
for i in range(size_matriz_quad):
  print(int(vectors_7[i][4 - i]), end=" ") #!!!

print("\n\nElementos acima da diagonal secundária:") # El. acima da diagonal secundária
for i in range(4):
  for j in range(3 - i, -1, -1): #!!!
    print(int(vectors_7[i][j]), end=" ")

print("\n\nElementos abaixo da diagonal secundária:") # El. abaixo da diagonal secundária
for i in range(1, size_matriz_quad):
  for j in range(size_matriz_quad - i, size_matriz_quad):
    print(int(vectors_7[i][j]), end=" ")
```

3. Entrar com valores para uma matriz A3 < 4. Gerar e imprimir uma matriz B que é o triplo da matriz A.

```python
num_vectors = 1
size_matriz_quad = 3
size_v = [3,3]

vectors_8 = create_vectors(num_vectors, size_v) # Criando matriz
vectors_8 = vectors_8[0]

print(f"Digite os valores da matriz A {size_matriz_quad}x{size_matriz_quad} (insira valores < que 4):")

for i in range(size_matriz_quad):
  print('\n')

  for j in range(size_matriz_quad):
    valor = float(input(f"Digite o valor de A[{i}][{j}]: "))

    while valor >= 4:
        print("Valor deve ser menor que 4. Tente novamente.")
        valor = float(input(f"Digite o valor de A[{i}][{j}]: "))

    vectors_8[i][j] = valor

B = 3 * vectors_8 # Multiplicar cada el. A por 3
print("\nMatriz B (triplo da matriz A):")
print(B)
```
